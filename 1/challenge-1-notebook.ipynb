<h4>Setting some common variables</h4>

```
file_type = "csv"

infer_schema = "false"
first_row_is_header = "true"
delimiter = ","
```

<h4>Loading GDP Data</h4>

```
#GDP Data
file_location = "/FileStore/tables/API_NY_GDP_MKTP_CD_DS2_en_csv_v2_4546417.csv"

rawGdpDf = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

renamedGdpDf = rawGdpDf.withColumnRenamed('Country Name', 'Country').withColumnRenamed('Country Code', 'CountryCode')

display(renamedDf)

renamedGdpDf.createOrReplaceTempView("gdpView")
```

<h4>Loading Population Data</h4>

```

file_location = "/FileStore/tables/API_SP_POP_TOTL_DS2_en_csv_v2_4558492.csv"

rawPopDf = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

renamedPopDf = rawPopDf.withColumnRenamed('Country Name', 'Country').withColumnRenamed('Country Code', 'CountryCode')

renamedPopDf.createOrReplaceTempView("popView")

display(renamedPopDf)
```

<h4>GBP Exchange Rate Data</h4>
Acquired Manually, converted to a dataframe

```

import pandas as pd

gbpData = [["2012", 0.631109], ["2013", 0.63955], ["2014", 0.607353], ["2015", 0.654441], ["2016", 0.740559], ["2017", 0.776691], ["2018", 0.750148], ["2019", 0.783652], ["2020", 0.779494], ["2021", 0.727434]]

gbpDf = pd.DataFrame(gbpData, columns=["Year", "GBPRate"])
gbpDf1 = spark.createDataFrame(gbpDf)
gbpDf1.createOrReplaceTempView("rateView")
```

<h4>Unpivoting data sets for past 10 years</h4>
For ease of manipulation into charts

```
countryList = "'United States', 'Japan', 'Canada', 'France', 'Italy', 'United Kingdom', 'Germany'"

query = "select Country, stack(10,'2012', `2012`, '2013',`2013`, '2014', `2014`, '2015', `2015`, '2016', `2016`,'2017', `2017`, '2018',`2018`, '2019', `2019`, '2020', `2020`, '2021', `2021`) as (Year, {0}) from `{1}` WHERE Country IN ({2})"

unPivotGdpDf = spark.sql(query.format("GDP","gdpView",countryList))
unPivotGdpDf.createOrReplaceTempView("gdpFilterView")

unPivotPopDf = spark.sql(query.format("Population","popView",countryList))
unPivotPopDf.createOrReplaceTempView("popFilterView")

```

<h4>Joining datasets</h4>

```
gdpGbpDf = spark.sql("SELECT g.Country, g.Year, g.GDP AS GDPUSD, (r.GBPRate * g.GDP) AS GDPGBP FROM gdpFilterView AS g JOIN rateView as r ON g.Year = r.Year")
gdpGbpDf.createOrReplaceTempView("gdpRateView")

joinedDf = spark.sql("SELECT g.Country, g.Year, g.GDPUSD, g.GDPGBP, p.Population, (g.GDPUSD/p.Population) AS gdpUSDPerCapita, (g.GDPGBP/p.Population) AS gdpGBPPerCapita  FROM gdpRateView AS g JOIN popFilterView AS p ON g.Country = p.Country AND g.Year = p.Year")
```

```
from pyspark.sql.functions import col

display(joinedDf.withColumn("GDPUSD",col("GDPUSD").cast("float")))
```
